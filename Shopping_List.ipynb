```python
import json
import os
#########################################
# Define function to view the list and total
def view_list(data):
    print("*" * 50)
    total = 0
    # Sorting the dictionary items alphabetically by key (item name)
    sorted_items = sorted(data.items())

    for key, value in sorted_items:
        total += value
        print(f"[ ] {key:20} ${value:<5.2f}")

    print("-" * 50)
    print(f"Your total is ${total:.2f}")
#########################################
# Define function to add an item to the list
def add_item(data):
     while True:
          item = input("Please enter the item name: type 'exit' to stop. ")

          # Exit condition
          if item.lower() == "exit":
               break

          try:
               price = float(input("Please enter the item price: "))
          except ValueError:
               print("Invalid price! Please enter a number.")
               continue

          data[item] = price

          # Save updated data to the file
          with open(FILENAME, "w") as file:
               json.dump(data, file, indent=4)
#########################################
# Define function to remove an item from the list
def remove_item(data):
     cnt = 0
     for key, value in data.items():
        print(f" {cnt} {key:20} ${value:<5.2f}")
        cnt +=1
     item_to_remove = int(input("Which item number to remove?: "))
     key_to_remove = list(data.keys())[item_to_remove]
     data.pop(key_to_remove)
#########################################
# Define function to edit an item from the list
def edit_item(data):
    cnt = 0
    for key, value in data.items():
        print(f" {cnt} {key:20} ${value:<5.2f}")
        cnt += 1

    try:
        item_to_edit = int(input("Which item number to edit?: "))
        key_to_edit = list(data.keys())[item_to_edit]
    except (ValueError, IndexError):
        print("Invalid selection! Please enter a valid item number.")
        return  # Exit function if input is invalid

    new_item_name = input("Please enter the new item name: ").strip()

    if not new_item_name:
        print("Item name cannot be empty!")
        return

    try:
        new_price = float(input("Please enter the new item price: "))
    except ValueError:
        print("Invalid price! Please enter a number.")
        return

    # Remove old key and insert the updated key-value pair
    del data[key_to_edit]
    data[new_item_name] = new_price

    # Save updated data to the file
    with open(FILENAME, "w") as file:
        json.dump(data, file, indent=4)

    print(f"Item '{key_to_edit}' updated to '{new_item_name}' with price ${new_price:.2f}")

#########################################
# Define the filename to store data
FILENAME = "shopping_list.json"
#########################################
# Load existing data if available
if os.path.exists(FILENAME):
     with open(FILENAME,"r") as file:
          try:
               data = json.load(file)
          except json.JSONDecoder:
               data = {}
else:
     data = {}
#########################################
# Display a welcome message on the screen
print()
print("Welcome!")
print("=" * 50)
#########################################
# Display options to the user
while True:
     print()
     print("0: View the existing list")
     print("1: Add new items")
     print("2: Remove an item from the list")
     print("3: Edit an item")
     print("4: Exit the program")
     print("-" * 25)
     choice = int(input("Enter your selection: "))
     if choice == 4:
          break
     elif choice == 0:
          view_list(data)
     elif choice == 1:
          add_item(data)
     elif choice == 2:
          remove_item(data)
     elif choice == 3:
          edit_item(data)
     else:
          print("Please enter a valid choice!")
```
