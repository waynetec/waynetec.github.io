```python
# Define function to return area
def area_calculation(length, width):
    return length * width

# Define function to return the perimeter
def perimeter_calculation(length, width):
    return (length * 2) + (width * 2)

# Define the cubic value
def room_cubic_calculation(length, width, height):
    if height <= 0:
        print("\tInvalid height: ")
        return None
    else:
        return length * width * height

# Define a dictionary to hold the room data
rooms = {}

# Get measurements from the user
def gui_function():
    while True:
        room_name = input("Enter a name for the room (Type 'quit' to stop): ").strip() or "quit"
        if room_name.lower() == 'quit':  # kill the program if user enters quit or doesn't enter anything for room name
            break

        try:
            room_length = float(input("Enter the room length: "))
            room_width = float(input("Enter the room width: "))
            room_height = float(input("Enter the room height: "))

            measurements = [room_length, room_width, room_height]
            rooms[room_name] = measurements
        except ValueError:
            print("\nInvalid input. Please enter numeric values for length, width, and height.\n")

    # Iterate through the keys and values and do calculations
    for key, values in rooms.items():
        print(f"{key}: ")
        print(f"\tMeasurements: {values[0]} x {values[1]} x {values[2]}")

        area = area_calculation(values[0], values[1])
        if area is not None:
            print(f"\tArea : {area:.2f}")

        peri = perimeter_calculation(values[0], values[1])
        if peri is not None:
            print(f"\tPerimeter: {peri:.2f}")

        cube = room_cubic_calculation(values[0], values[1], values[2])
        if cube is not None:
            print(f"\tCubic: {cube:.2f}")

gui_function()

```
